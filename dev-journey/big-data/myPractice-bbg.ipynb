```python
df = pd.read_csv('samples/data.csv', encoding='cp949')
df['sum'] = df['쇼핑1월'] + df['쇼핑2월'] + df['쇼핑3월']
df['avg'] = df[['쇼핑1월', '쇼핑2월', '쇼핑3월']].apply(np.mean, axis=1)    # 여기 axis=1은 행별 연산이다
df[['sum', 'avg']]                                                       # df['avg'] = df['sum']/3 도 가능
```

```python
df = pd.read_csv('samples/usedcars.csv', encoding='cp949')
dfnew = df[['maker', 'fuel']]
display(dfnew.value_counts())                                       # 평범한 value_counts
display(pd.crosstab(dfnew['maker'], dfnew['fuel'], margins=True))   # 크로스 테이블 (교차테이블)
dfnew.value_counts(normalize=True) * 100                            # 백분율로 상대빈도 
```

```python
# 변동계수 (Coefficient of Variation) CV = std / mean 으로 측정 단위가 다른 자료를 비교할 때 쓰임.
# df['열이름'].quantile(0.25) == np.percentile(df['열이름'], 25) == df['열이름'].describe()['25%']
from scipy.stats import skew, kurtosis
p(skew(df['price']))
p(kurtosis(df['price']))
```
