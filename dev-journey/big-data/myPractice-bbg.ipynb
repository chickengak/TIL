```python
# 상관관계 분석은 선형적 관계만 알 수 있고, 함수적 관계는 회귀분석을 해야 한다.
# 표본 상관계수 = Cov(x, y) / (std(x) * std(y))
# 피어슨 상관계수: 등간척도 자료 사이의 관계를 알아보기 위해 사용되며, 연속형 변수 사이의 선형적 관계를 나타낸다.
# 켄달 상관계수: 범주형 자료(서열척도)인 두 변수 사이의 상관관계를 측정. 비모수적 통계분석 기법으로서 순위 상관계수의 한 종류.
# 스피어만 상관계수: 범주형 자료(서열척도)인 두 변수 사이의 상관관계를 측정.
```

상관계수의 성질
- 상관계수는 -1 <= r <= 1 사이에 있다.
- 0 보다 크면 양의 (선형) 상관관계, 0 보다 작으면 음의 (선형) 상관관계

```python
df = pd.read_csv('samples/women.csv')
print('-- 공분산 --', res:=df[['height', 'weight']].cov(),  res.iloc[0, 1], '',sep='\n')
print('-- 피어슨 상관관계 --', res:=df[['height', 'weight']].corr(method='pearson'),  res.iloc[0, 1], '',sep='\n')
print('-- 스피어만 상관관계 --', res:=df[['height', 'weight']].corr(method='spearman'),  res.iloc[0, 1], '',sep='\n')
print('-- 켄달 상관관계 --', res:=df[['height', 'weight']].corr(method='kendall'),  res.iloc[0, 1], '',sep='\n')
```

```python
# 위의 키와 몸무게의 상관관계가 유의한 지를 가설 검정해보자.
from scipy import stats

print(stats.pearsonr(df['height'], df['weight']))       # (상관계수, p-value)
print(res:= stats.ttest_ind(df['height'], df['weight']))      # (T-검정통계량, p-value)
print(res[1] < 0.01)    # True이므로 귀무가설 기각, 대립가설 채택
```

```python
df = pd.read_csv('samples/mtcars.csv')[['mpg', 'wt', 'qsec', 'disp', 'drat']]
print('-- 상관계수 --')
display(df.corr())
print('-- 검정 -- ')
dfres = pd.DataFrame(np.zeros((df.shape[1], df.shape[1])), columns=df.columns)
for i in range(0, len(df.columns)):
    col1 = df.columns[i]
    for j in range(i+1, len(df.columns)):
        col2 = df.columns[j]
        res = stats.ttest_ind(df[col1], df[col2])
        print(col1, '\t과', col2, ' 의 검정통계량과 p-value: \t', res)
        dfres.iloc[i, j] = res[1]
display(dfres)
```

```python
### 단순 회귀 분석
from sklearn.linear_model import LinearRegression

df = pd.read_csv('samples/women.csv')
x = df['height'].values.reshape(-1, 1)
y = df['weight']
fit = LinearRegression().fit(x, y)
print(fit.intercept_, fit.coef_)    # y절편, 기울기
residual = y - fit.predict(x)
print(residual.mean())              # 잔차의 평균
plt.figure(figsize=(8,4))
plt.scatter(x, y)
plt.plot(x, fit.predict(x), color='red')
plt.xlabel('Height'); plt.ylabel('Weight')
plt.show()
```

### 로지스틱 회귀
- sklearn.linear_model.LogisticRegression(C=20, max_iter=100)
- C=20은 릿지회귀 L2 규제값. C값이 클수록 정규화는 약해지고 C값이 작아지면 정규화가 강해짐. 기본값은 1
- max_iter는 학습과정에서 최대 반복횟수를 제어함 (기본값 100). 여기에선 경사하강법으로 학습하는데, max_iter값만큼 경사하강법을 반복한다고 보면 됨

```python
from sklearn.datasets import load_iris
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

iris = load_iris()
iris_X = pd.DataFrame(iris['data'], columns=iris['feature_names'])
iris_Y = iris['target']

scaler = StandardScaler()
X = scaler.fit_transform(iris_X)

model = LogisticRegression(C=20)
model.fit(X, iris_Y)

print('정확도: ', model.score(X, iris_Y)*100)
print('coef_ \n', model.coef_)
print('coef_ 모양: ', model.coef_.shape, '각 클래스(3개) 마다 특징(4개)의 계수')
print('z 절편값 \n', model.intercept_)
print('z 절편값 모양: ', model.intercept_.shape, '각 클래스(3개) 마다 절편값')
print()
print(model.decision_function(X[:5]).round(3))   # 각 클래스일 시그모이드 독립변수 값
print(model.predict_proba(X[:5]).round(3))       # 각 클래스일 예측 확률. 시그모이드를 거친 값
```

## 의사결정 나무

```python
### iris 데이터
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree

iris = load_iris()
iris_X = pd.DataFrame(iris['data'], columns=iris['feature_names'])
iris_Y = pd.DataFrame(iris['target'], columns=['target']).apply(lambda x: iris['target_names'][x])
train_X, val_X, train_Y, val_Y = train_test_split(iris_X, iris_Y, train_size=0.7, random_state=55)

scaler = StandardScaler()
train_X_scaled = scaler.fit_transform(train_X)
val_X_scaled = scaler.transform(val_X)

model = DecisionTreeClassifier(random_state=55)     # max_depth=3 넣으면, 오히려 성능 줄어듦. 그리고 특성 중요도도 바뀜.
model.fit(train_X_scaled, train_Y)

print('정확도: 훈련 -', model.score(train_X_scaled, train_Y), ' 검증 -', model.score(val_X_scaled, val_Y))
print('특성 중요도:', model.feature_importances_)
plt.figure(figsize=(16,4))
plt.subplot(1,2,1)
plot_tree(model)
plt.subplot(1,2,2)
plot_tree(model, max_depth=3, filled=True, feature_names=iris['feature_names']) # max_depth=3: 루트노드를 제외하고 세개의 노드를 더확장하여 작성
plt.show()                      # filled=True: 클래스별로 노드 색상 지정    
```

```python
### 검증하기
# 테이블로 검증
table = pd.DataFrame(np.zeros((45, 2)), columns=('actual', 'predict'))
table['actual'] = val_Y['target'].to_list()
table['predict'] = model.predict(val_X_scaled)
table['compare'] = table['actual'] == table['predict']
display(table.head())
print(table['compare'].sum() / len(table), ' <- 정확도')

# 지표
from sklearn.metrics import classification_report, confusion_matrix

conf = confusion_matrix(table['actual'], table['predict'])
print('혼동행렬표 \n', conf)
print('분류 분석 모형 성능평가 지표 \n', classification_report(table['actual'], table['predict']))
```

```python
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

iris = load_iris()
iris_X = pd.DataFrame(iris['data'], columns=iris['feature_names'])
iris_Y = pd.DataFrame(iris['target'], columns=['target']).apply(lambda x: iris['target_names'][x])
train_X, val_X, train_Y, val_Y = train_test_split(iris_X, iris_Y, train_size=0.7, random_state=42)

scaler = StandardScaler()
train_X_scaled = scaler.fit_transform(train_X)
val_X_scaled = scaler.transform(val_X)

model = MLPClassifier(hidden_layer_sizes=(100, 50), max_iter=500, random_state=42)
model.fit(train_X_scaled, train_Y)

predicted = model.predict(val_X_scaled)
print("정확도: ", model.score(val_X_scaled, val_Y), accuracy_score(val_Y, predicted))
print('혼동행렬표: \n', confusion_matrix(val_Y, predicted))
print('성능 평가 지표: \n', classification_report(val_Y, predicted))
```

## SVM - 서포트 벡터 머신

- SVM은 예측 정확도가 비교적 높고, 사용이 쉽다는 장점이 있으며, 신경망에 비해 과대적합이 덜하다는 특징이 있다.
- 반면, 커널 함수의 적절한 파라미터를 설정하기 위해서는, 여러 번 테스트해야만 최적화 모형을 만들 수 있기 때문에, 모형 구축 시간이 오래 걸린다는 단점이 있고, 결과에 대한 설명력이 다른 기법들에 비해 다소 떨어진다는 특징이 있다.

```python
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

iris = load_iris()
iris_X = pd.DataFrame(iris['data'], columns=iris['feature_names'])
iris_Y = pd.DataFrame(iris['target'], columns=['target']).apply(lambda x: iris['target_names'][x])
train_X, val_X, train_Y, val_Y = train_test_split(iris_X, iris_Y, train_size=0.7, random_state=42)

scaler = StandardScaler()
train_X_scaled = scaler.fit_transform(train_X)
val_X_scaled = scaler.transform(val_X)

model = SVC(kernel='linear', C=10, gamma=0.1, random_state=42)
model.fit(train_X_scaled, train_Y)

predicted = model.predict(val_X_scaled)
print('정확도:  훈련 -', accuracy_score(train_Y, model.predict(train_X_scaled)),' 검증 -', accuracy_score(val_Y, predicted))
print('혼동행렬: \n', confusion_matrix(val_Y, predicted))
print('성능 평가 지표: \n', classification_report(val_Y, predicted))
```

```python
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

iris = load_iris()
iris_X = pd.DataFrame(iris['data'], columns=iris['feature_names'])
iris_Y = pd.DataFrame(iris['target'], columns=['target']).apply(lambda x: iris['target_names'][x])
train_X, val_X, train_Y, val_Y = train_test_split(iris_X, iris_Y, train_size=0.7, random_state=42)

scaler = StandardScaler()
train_X_scaled = scaler.fit_transform(train_X)
val_X_scaled = scaler.transform(val_X)

model = SVC(random_state=42)
param_grid = {
    'kernel': ['poly', 'linear', 'sigmoid'],
    'C': [0.1, 1, 10, 100],
    'gamma': [0.1, 0.01, 0.001]
}

gridSearch = GridSearchCV(model, param_grid, cv=5, n_jobs=1)
gridSearch.fit(train_X_scaled, train_Y)

print('최적의 파라미터: ', gridSearch.best_params_)
bestModel = gridSearch.best_estimator_
accuracy = bestModel.score(val_X_scaled, val_Y)
print('정확도: ', accuracy)
```

## 베이지안 기법

- 가우시안 나이브 베이즈는 연속적인 수치형 데이터를 다루는데 효과적. 주어진 레이블에 속할 확률을 계산하기 위해 가우시안 분포를 사용.

```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import CategoricalNB
from sklearn.metrics import accuracy_score

df = pd.read_csv('samples/data.csv', encoding='cp949')
df['gender'] = df['성별'].apply(lambda x: 1 if str(x) == '남자' else 0 )
df['house'] = df['주거지역'].map({'소도시': 0, '중도시': 1, '대도시': 2})
df['like'] = df['쿠폰선호도'].map({'예': 1, '아니오': 0})
display(df.head(2))
train_X, val_X, train_Y, val_Y = train_test_split(df[['gender', 'house']], df['like'], train_size=0.7, random_state=42)

model = CategoricalNB()#alpha=0.8)
model.fit(train_X, train_Y)

predicted = model.predict(val_X)
print('정확도: 훈련 -', accuracy_score(train_Y, model.predict(train_X)),'검정 -', accuracy_score(val_Y, predicted))
print('파라미터: ', model.get_params())
```

```python
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, auc, f1_score, accuracy_score, precision_score, recall_score

print('혼동행렬표: \n', confusion_matrix(val_Y, predicted))
print('성능 평가 지표: \n', classification_report(val_Y, predicted))
print('F1 score: ', f1_score(val_Y, predicted))
print('ACC, Precision, Recall: ', accuracy_score(val_Y, predicted), precision_score(val_Y, predicted), recall_score(val_Y, predicted))
print('ROC Curve: ')
fpr, tpr, thresholds = roc_curve(val_Y, predicted)
plt.figure(figsize=(3,2)); plt.plot(fpr, tpr); plt.show()
print('auc: ', auc(fpr, tpr))
```

- Boosting 부스팅
    - 약분류기를 강분류기로

```python
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection  import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

iris = load_iris()
iris_X = pd.DataFrame(iris['data'], columns=iris['feature_names'])
iris_Y = pd.DataFrame(iris['target'], columns=['target']).apply(lambda x: iris['target_names'][x])
train_X, val_X, train_Y, val_Y = train_test_split(iris_X, iris_Y, train_size=0.7, random_state=42)

scaler = StandardScaler()
train_X_scaled = scaler.fit_transform(train_X)
val_X_scaled = scaler.transform(val_X)

model = RandomForestClassifier(random_state=42)
model.fit(train_X_scaled, train_Y)

predicted = model.predict(val_X_scaled)
print('정확도: ', accuracy_score(val_Y, predicted))
print('특성 중요도: ', model.feature_importances_)
print('혼동행렬표: \n', confusion_matrix(val_Y, predicted))
print('성능 평가 지표: \n', classification_report(val_Y, predicted))
```

```python
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection  import train_test_split, GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

iris = load_iris()
iris_X = pd.DataFrame(iris['data'], columns=iris['feature_names'])
iris_Y = pd.DataFrame(iris['target'], columns=['target']).apply(lambda x: iris['target_names'][x])
train_X, val_X, train_Y, val_Y = train_test_split(iris_X, iris_Y, train_size=0.7, random_state=42)

scaler = StandardScaler()
train_X_scaled = scaler.fit_transform(train_X)
val_X_scaled = scaler.transform(val_X)

model = RandomForestClassifier(random_state=42)
grid_params = {
    'n_estimators': [50, 100, 200],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 3, 4],
    'min_samples_leaf': [1, 2, 3]
}
gridSearch = GridSearchCV(model, grid_params)
gridSearch.fit(train_X_scaled, train_Y)

bestModel = gridSearch.best_estimator_
predicted = bestModel.predict(val_X_scaled)
print('정확도:  훈련 -', gridSearch.best_score_, ' 검증 -', accuracy_score(val_Y, predicted))
print('최적의 파라미터: ', gridSearch.best_params_)
print('혼동행렬표: \n', confusion_matrix(val_Y, predicted))
print('성능 평가 지표: \n', classification_report(val_Y, predicted))
```

```python
# 랜덤 포레스트는 학습과 검증으로 나누지 않고, OOB(Out Of Bag, 부트스트랩 새플에 포함되지 않는 남은 샘플)를 이용한 학습도 가능하다.

model = RandomForestClassifier(oob_score=True, max_depth=None, min_samples_leaf=1, min_samples_split=2, n_estimators=100, random_state=42)
model.fit(train_X_scaled, train_Y)

print('OOB를 이용한 검정 정확도: ', model.oob_score_)
print('훈련 정확도: ', accuracy_score(train_Y, model.predict(train_X_scaled)))
```
