```python
# 상관관계 분석은 선형적 관계만 알 수 있고, 함수적 관계는 회귀분석을 해야 한다.
# 표본 상관계수 = Cov(x, y) / (std(x) * std(y))
# 피어슨 상관계수: 등간척도 자료 사이의 관계를 알아보기 위해 사용되며, 연속형 변수 사이의 선형적 관계를 나타낸다.
# 켄달 상관계수: 범주형 자료(서열척도)인 두 변수 사이의 상관관계를 측정. 비모수적 통계분석 기법으로서 순위 상관계수의 한 종류.
# 스피어만 상관계수: 범주형 자료(서열척도)인 두 변수 사이의 상관관계를 측정.
```

상관계수의 성질
- 상관계수는 -1 <= r <= 1 사이에 있다.
- 0 보다 크면 양의 (선형) 상관관계, 0 보다 작으면 음의 (선형) 상관관계

```python
df = pd.read_csv('samples/women.csv')
print('-- 공분산 --', res:=df[['height', 'weight']].cov(),  res.iloc[0, 1], '',sep='\n')
print('-- 피어슨 상관관계 --', res:=df[['height', 'weight']].corr(method='pearson'),  res.iloc[0, 1], '',sep='\n')
print('-- 스피어만 상관관계 --', res:=df[['height', 'weight']].corr(method='spearman'),  res.iloc[0, 1], '',sep='\n')
print('-- 켄달 상관관계 --', res:=df[['height', 'weight']].corr(method='kendall'),  res.iloc[0, 1], '',sep='\n')
```

```python
# 위의 키와 몸무게의 상관관계가 유의한 지를 가설 검정해보자.
from scipy import stats

print(stats.pearsonr(df['height'], df['weight']))       # (상관계수, p-value)
print(res:= stats.ttest_ind(df['height'], df['weight']))      # (T-검정통계량, p-value)
print(res[1] < 0.01)    # True이므로 귀무가설 기각, 대립가설 채택
```

```python
df = pd.read_csv('samples/mtcars.csv')[['mpg', 'wt', 'qsec', 'disp', 'drat']]
print('-- 상관계수 --')
display(df.corr())
print('-- 검정 -- ')
dfres = pd.DataFrame(np.zeros((df.shape[1], df.shape[1])), columns=df.columns)
for i in range(0, len(df.columns)):
    col1 = df.columns[i]
    for j in range(i+1, len(df.columns)):
        col2 = df.columns[j]
        res = stats.ttest_ind(df[col1], df[col2])
        print(col1, '\t과', col2, ' 의 검정통계량과 p-value: \t', res)
        dfres.iloc[i, j] = res[1]
display(dfres)
```

### 로지스틱 회귀
- sklearn.linear_model.LogisticRegression(C=20, max_iter=100)
- C=20은 릿지회귀 L2 규제값. C값이 클수록 정규화는 약해지고 C값이 작아지면 정규화가 강해짐. 기본값은 1
- max_iter는 학습과정에서 최대 반복횟수를 제어함 (기본값 100). 여기에선 경사하강법으로 학습하는데, max_iter값만큼 경사하강법을 반복한다고 보면 됨

```python
from sklearn.datasets import load_iris
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

iris = load_iris()
iris_X = pd.DataFrame(iris['data'], columns=iris['feature_names'])
iris_Y = iris['target']

scaler = StandardScaler()
X = scaler.fit_transform(iris_X)

model = LogisticRegression(C=20)
model.fit(X, iris_Y)

print('정확도: ', model.score(X, iris_Y)*100)
print('coef_ \n', model.coef_)
print('coef_ 모양: ', model.coef_.shape, '각 클래스(3개) 마다 특징(4개)의 계수')
print('z 절편값 \n', model.intercept_)
print('z 절편값 모양: ', model.intercept_.shape, '각 클래스(3개) 마다 절편값')
print()
print(model.decision_function(X[:5]).round(3))   # 각 클래스일 시그모이드 독립변수 값
print(model.predict_proba(X[:5]).round(3))       # 각 클래스일 예측 확률. 시그모이드를 거친 값
```

```python
### iris 데이터
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree

iris = load_iris()
iris_X = pd.DataFrame(iris['data'], columns=iris['feature_names'])
iris_Y = pd.DataFrame(iris['target'], columns=['target']).apply(lambda x: iris['target_names'][x])
train_X, val_X, train_Y, val_Y = train_test_split(iris_X, iris_Y, train_size=0.7, random_state=55)

scaler = StandardScaler()
train_X_scaled = scaler.fit_transform(train_X)
val_X_scaled = scaler.transform(val_X)

model = DecisionTreeClassifier(random_state=55)     # max_depth=3 넣으면, 오히려 성능 줄어듦. 그리고 특성 중요도도 바뀜.
model.fit(train_X_scaled, train_Y)

print('정확도: 훈련 -', model.score(train_X_scaled, train_Y), ' 검증 -', model.score(val_X_scaled, val_Y))
print('특성 중요도:', model.feature_importances_)
plt.figure(figsize=(16,4))
plt.subplot(1,2,1)
plot_tree(model)
plt.subplot(1,2,2)
plot_tree(model, max_depth=3, filled=True, feature_names=iris['feature_names']) # max_depth=3: 루트노드를 제외하고 세개의 노드를 더확장하여 작성
plt.show()                      # filled=True: 클래스별로 노드 색상 지정    
```

```python
### 검증하기
# 테이블로 검증
table = pd.DataFrame(np.zeros((45, 2)), columns=('actual', 'predict'))
table['actual'] = val_Y['target'].to_list()
table['predict'] = model.predict(val_X_scaled)
table['compare'] = table['actual'] == table['predict']
display(table.head())
print(table['compare'].sum() / len(table), ' <- 정확도')

# 지표
from sklearn.metrics import classification_report, confusion_matrix

conf = confusion_matrix(table['actual'], table['predict'])
print('혼동행렬표 \n', conf)
print('분류 분석 모형 성능평가 지표 \n', classification_report(table['actual'], table['predict']))
```

## SVM - 서포트 벡터 머신

- SVM은 예측 정확도가 비교적 높고, 사용이 쉽다는 장점이 있으며, 신경망에 비해 과대적합이 덜하다는 특징이 있다.
- 반면, 커널 함수의 적절한 파라미터를 설정하기 위해서는, 여러 번 테스트해야만 최적화 모형을 만들 수 있기 때문에, 모형 구축 시간이 오래 걸린다는 단점이 있고, 결과에 대한 설명력이 다른 기법들에 비해 다소 떨어진다는 특징이 있다.

```python
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

iris = load_iris()
iris_X = pd.DataFrame(iris['data'], columns=iris['feature_names'])
iris_Y = pd.DataFrame(iris['target'], columns=['target']).apply(lambda x: iris['target_names'][x])
train_X, val_X, train_Y, val_Y = train_test_split(iris_X, iris_Y, train_size=0.7, random_state=42)

scaler = StandardScaler()
train_X_scaled = scaler.fit_transform(train_X)
val_X_scaled = scaler.transform(val_X)

model = SVC(kernel='linear', C=10, gamma=0.1, random_state=42)
model.fit(train_X_scaled, train_Y)

predicted = model.predict(val_X_scaled)
print('정확도:  훈련 -', accuracy_score(train_Y, model.predict(train_X_scaled)),' 검증 -', accuracy_score(val_Y, predicted))
print('혼동행렬: \n', confusion_matrix(val_Y, predicted))
print('성능 평가 지표: \n', classification_report(val_Y, predicted))
```

```python
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

iris = load_iris()
iris_X = pd.DataFrame(iris['data'], columns=iris['feature_names'])
iris_Y = pd.DataFrame(iris['target'], columns=['target']).apply(lambda x: iris['target_names'][x])
train_X, val_X, train_Y, val_Y = train_test_split(iris_X, iris_Y, train_size=0.7, random_state=42)

scaler = StandardScaler()
train_X_scaled = scaler.fit_transform(train_X)
val_X_scaled = scaler.transform(val_X)

model = SVC(random_state=42)
param_grid = {
    'kernel': ['poly', 'linear', 'sigmoid'],
    'C': [0.1, 1, 10, 100],
    'gamma': [0.1, 0.01, 0.001]
}

gridSearch = GridSearchCV(model, param_grid, cv=5, n_jobs=1)
gridSearch.fit(train_X_scaled, train_Y)

print('최적의 파라미터: ', gridSearch.best_params_)
bestModel = gridSearch.best_estimator_
accuracy = bestModel.score(val_X_scaled, val_Y)
print('정확도: ', accuracy)
```

## 베이지안 기법

# 군집, K-NN 분류 분석

#### **▶ 군집 ◀**

- 각 개체들에 대한 사전 지식 없이 유사성(Similarity, 유사도, 거리 등) 혹은 상이성(이질성)에 근거하여 군집을 분류하는 것으로서, 독립변수 값만으로 새로운 개체를 예측한다.
- 군집분석은 크게 계층적 군집분석과 비계층적(분할적) 군집분석으로 나뉨.
- 계층적 군집분석 (Hierarchical Clustering Analysis)
    - 주로 자료의 크기가 작을 때 사용
    - 개별 대상들 사이의 거리에 의하여 가장 가까이에 있는 대상들로부터 시작하여 결합해 나아감으로써 남 모양의 계층구조를 형성함.
    - 계층적 군집분석에는 주로 덴드로그램(Dendrogram, 개체들이 결합되는 순서를 나타내는 트리 형태의 구조)을 이용하여 군집이 형성되는 과정을 명확히 파악할 수 있다.
- 비계층적 군집분석 (Non-hierachical 또는 Partitional Clustering Analysis)
    - 구하고자 하는 군집의 수를 미리 정함. 그리고 임의의 군집 중심에서 가장 가까운 개체를 하나씩 포함해 나아가는 방식.
    - 설정하는 군집 수와 군집을 형성하기 위한 초기값에 따라 군집의 결과가 달라지는 단점이 존재
    - 자료 크기가 크거나 하나의 개체가 여러 군집에 포함되는(중복 비계층적) 경우 사용


```python
### 계층적 군집분석
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from scipy.cluster.hierarchy import linkage, dendrogram, fcluster

iris = load_iris()
iris_X = pd.DataFrame(iris['data'], columns=iris['feature_names'])
iris_Y = pd.DataFrame(iris['target'], columns=['target']).apply(lambda x: iris['target_names'][x])

scaler = StandardScaler()
iris_X_scaled = scaler.fit_transform(iris_X)

clusters = linkage(iris_X_scaled, method='complete')     # complete linkage를 이용한 계층적 군집 분석

cut = fcluster(clusters, t=3, criterion='distance')
print("군집결과: method='complete' \n", cut)

plt.figure(figsize=(16, 4))
dendrogram(clusters)                # 덴드로그램 작성
plt.title("method='complete'"); plt.xlabel('Data Points'); plt.ylabel('Distance'); plt.show()

clusters_avg = linkage(iris_X_scaled, method='average')     # 평균 연결법
# print("군집결과: method='average' \n", fcluster(clusters_avg, t=3, criterion='distance'))
plt.figure(figsize=(16, 4))
dendrogram(clusters_avg)
plt.title("method='average'"); plt.xlabel('Data Points'); plt.ylabel('Distance'); plt.show()

clusters_centroid = linkage(iris_X_scaled, method='centroid')   # 중심 연결법
# print("군집결과: method='average' \n", fcluster(clusters_centroid, t=3, criterion='distance'))
plt.figure(figsize=(16, 4))
dendrogram(clusters_centroid)
plt.title("method='centroid'"); plt.xlabel('Data Points'); plt.ylabel('Distance'); plt.show()

clusters_ward = linkage(iris_X_scaled, method='ward')   # 와드 연결법
# print("군집결과: method='average' \n", fcluster(clusters_ward, t=3, criterion='distance'))
plt.figure(figsize=(16, 4))
dendrogram(clusters_ward)
plt.title("method='ward'"); plt.xlabel('Data Points'); plt.ylabel('Distance'); plt.show()

```

```python
### 비계층적 군집분석 - K-Means Clustering K-평균 군집
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA

iris = load_iris()
iris_X = pd.DataFrame(iris['data'], columns=iris['feature_names'])
iris_Y = pd.DataFrame(iris['target'], columns=['target']).apply(lambda x: iris['target_names'][x])

scaler = StandardScaler()
iris_X_scaled = scaler.fit_transform(iris_X)

pca = PCA(2)    # 주성분분석. n_components=2
xpca = pca.fit_transform(iris_X_scaled)

model = KMeans(3, random_state=42)  # 군집 수 3으로 설정. n_clusters=3
model.fit(iris_X_scaled)

labels = model.labels_
print('KMeans 결과 라벨링: \n', labels)

plt.figure(figsize=(10, 5))
colors = ['red', 'green', 'blue']
for i in range(3):
    plt.scatter(xpca[labels == i, 0], xpca[labels == i, 1], c=colors[i], label=f"Cluster {i+1}")
plt.scatter(model.cluster_centers_[:,0], model.cluster_centers_[:,1], s=300, c='yellow', label='Centroids')
plt.xlabel('PCA Dimension #1'); plt.ylabel('PCA Dimension #2'); plt.legend(); plt.show()
```

#### **▶ K-NN 분류 분석 ◀**
- 새로운 데이터가 들어왔을 때, 기존 데이터의 그룹 중 어떤 그룹에 속하는지 분류함.
- 비교적 간단해서 구현이 쉬움. 수치형 자료의 분류 작업에서 성능이 우수.
- 데이터 양이 많으면 분류 속도가 기하급수적으로 느려지는 단점.
- k의 값은 보통 홀수. k의 값이 커지면 이상적인 분류를 할 수 있으나, 너무 많은 분류 작업을 수행해야 해서 분류 자체가 불가능한 경우가 발생할 수 있다.
- 즉, k의 값이 작으면 지역적인 패턴에 민감하지만 노이즈에도 민감해지는 반면, k의 값이 커지면 전역전인 패턴을 반영하지만 덜민감한 모델이 된다.

# 추론통계
추론통계란 모집단에서 표본을 수집하여 표본의 특성을 파악한 후, 표본의 특성을 모집단의 특성으로 일반화할 수 있는지 여부를 판단하여, 모집단의 특성인 모수를 추정하는 통계 분석 방법이다.

- $\chi^2$ 분포

#### **▶ 평균 차이 분석 ◀**
- 평균차이 분석은 두 집단 간의 평균을 비교하여, 평균의 차이가 통계적으로 유의한지 검정함.
- t-test를 이용함. (t-검정)
- 일반적으로, 모분산을 모른다면, 표본분산을 이용해서 검정한다.
- 독립표본 t-검정, 대응표본 t-검정으로 나뉨.

```python
### 성별에 따른 쇼핑 금액의 차이가 있는지 검정
from scipy import stats

df = pd.read_csv('samples/data.csv', encoding='cp949')
display(df.head())
print(df['성별'].unique())

### 검정통계량과 유의확률
t, pvalue = stats.ttest_ind(df.loc[df['성별'] == '남자']['쇼핑액'], df.loc[df['성별'] == '여자']['쇼핑액'], equal_var=False)
# equal_var=False <- 등분산성을 가정하지 않음. / alternative='two_sided' <- 기본값. 양측검정.
print('t-검정통계량: ', t)
print('pvalue(유의확률): ', pvalue, pvalue < 0.05)      # 귀무가설을 기각할 수 없으므로, 성별에 따른 쇼핑 금액 차이는 없다.

### 모평균의 차에 대한 95% 신뢰구간
meandiff = (male:=df.loc[df['성별'] == '남자']['쇼핑액']).mean() - (female:=df.loc[df['성별'] == '여자']['쇼핑액']).mean()
alpha = 0.05        # 유의수준
dof = len(df) - 2   # 자유도 degree of freedom
std_error = np.sqrt(np.var(male)/len(male) + np.var(female)/len(female))
margin_error = stats.t.ppf(1 - alpha/2, dof) * std_error
conf_interval = (meandiff - margin_error, meandiff + margin_error)
print('\n남자 쇼핑액과 여자 쇼핑액의 차이에 대한 신뢰구간(95%) = 모평균의 차에 대한 95% 신뢰구간: \n', conf_interval)

### 남자, 여자 각각에 따른 95% 신뢰구간
dof = len(male) - 1
std_error = np.sqrt(np.var(male)/len(male)) # == np.std(male)/np.sqrt(len(male))
margin_error_m = stats.t.ppf(1 - alpha/2, dof) * std_error
conf_interval = (male.mean() - margin_error_m, male.mean() + margin_error_m)
print('남자 쇼핑엑에 대한 95% 신뢰 구간: ', conf_interval)

dof = len(female) - 1
std_error = np.sqrt(np.var(female)/len(female))
margin_error_f = stats.t.ppf(1 - alpha/2, dof) * std_error
conf_interval = (female.mean() - margin_error_f, female.mean() + margin_error_f)
print('여성 쇼핑액에 대한 95% 신뢰 구간: ', conf_interval)

plt.figure(figsize=(3,4)); plt.bar(['male', 'female'], [male.mean(), female.mean()], color=['blue','red'], yerr=[margin_error_m, margin_error_f], alpha=0.3, capsize=10)
plt.show()      # 성별에 따른 쇼핑액의 분포는 비슷하나, 여자의 쇼핑액에 대한 신뢰구간이 다소 넓은 것을 알 수 있다.
```

```python
### 여자의 쇼핑액이 많은가를 판단? (적은가를 판단한거 아님?) (이게 대응표본 t-검정의 단측검정?)
from scipy import stats

df = pd.read_csv('samples/data.csv', encoding='cp949')
male = df.loc[df['성별'] == '남자']['쇼핑액']
female = df.loc[df['성별'] == '여자']['쇼핑액']

t, pvalue = stats.ttest_ind(male, female, equal_var=False, alternative='less')  # 귀무가설 male >= female , 대립가설 male < female
print('t-검정통계량: ', t)
print('pvalue(유의확률): ', pvalue, pvalue < 0.05)      # 귀무가설 채택. 남자의 쇼핑액이 여자보다 낮다고 볼 수 없다.
```

```python
### 남자의 쇼핑액이 여자의 쇼핑액보다 많은가?
from scipy import stats

df = pd.read_csv('samples/data.csv', encoding='cp949')
male = df.loc[df['성별'] == '남자']['쇼핑액']
female = df.loc[df['성별'] == '여자']['쇼핑액']

t, pvalue = stats.ttest_ind(male, female, equal_var=False, alternative='greater')   # 대립가설: male > female
print('t-검정통계량: ', t)
print('pvalue(유의확률): ', pvalue, pvalue < 0.05)
```

#### **▶ 비율 차이 분석 ◀**
- 어떤 항목에 대해 두 집단 사이의 비율에 차이가 있는지 검정
- 평균 차이 분석은 수치형 자료의 차이를 검정했다면, 비율 차이 분석은 명목형 자료의 차이를 검정한다.

```python
### 성별에 따라 쿠폰선호도의 비율 차이가 있는지 검정
from scipy import stats

df = pd.read_csv('samples/data.csv', encoding='cp949')
print(df['성별'].unique(), df['쿠폰선호도'].unique())
male = df.loc[df['성별'] == '남자']['쿠폰선호도']
female = df.loc[df['성별'] == '여자']['쿠폰선호도']

observed = [[male.value_counts()['예'], female.value_counts()['예']], [male.value_counts()['아니오'], female.value_counts()['아니오']]]
            # 남자-쿠폰선호, 여자-쿠폰선호, 남자-쿠폰미선호, 여자-쿠폰미선호
print(observed)
chi, pvalue, dof, expect = stats.chi2_contingency(observed)     # 카이제곱 검정
print('카이제곱 검정 통계량: ', chi)
print('pvalue(유의확률): ', pvalue, pvalue < 0.05)      # 대립가설 채택, 귀무가설 기각. 성별에 따른 쿠폰선호도 차이는 있다.
print('기대 빈도수: \n', expect)
```
