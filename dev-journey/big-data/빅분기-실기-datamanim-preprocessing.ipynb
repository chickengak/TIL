```python
# Q57. Income_Category의 카테고리를 map 함수를 이용하여 다음과 같이 변경하여 newIncome 컬럼에 매핑하라
"""
Unknown : N
Less than $40K : a
$40K - $60K : b
$60K - $80K : c
$80K - $120K : d
$120K +’ : e
"""

mydict = {'Unknown' : 'N', 'Less than $40K' : 'a', '$40K - $60K' : 'b', '$60K - $80K' : 'c', '$80K - $120K' : 'd', '$120K +' : 'e'}
df['newIncome'] = df['Income_Category'].map(mydict)
df
```

```python
# Q58. Income_Category의 카테고리를 apply 함수를 이용하여 다음과 같이 변경하여 newIncome 컬럼에 매핑하라
"""
Unknown : N
Less than $40K : a
$40K - $60K : b
$60K - $80K : c
$80K - $120K : d
$120K +’ : e
"""

def changeCategory(x):
    if x =='Unknown':
        return 'N'
    elif x =='Less than $40K':
        return 'a'
    elif x =='$40K - $60K':   
        return 'b'
    elif x =='$60K - $80K':    
        return 'c'
    elif x =='$80K - $120K':   
        return 'd'
    elif x =='$120K +' :     
        return 'e'
df['newIncome'] = df['Income_Category'].apply(changeCategory)
df
```

```python
# Q59. Customer_Age의 값을 이용하여 나이 구간을 AgeState 컬럼으로 정의하라.
#  (0~9 : 0 , 10~19 :10 , 20~29 :20 …) 그리고 각 구간의 빈도수를 출력하라

def addColumnAgeState(i):
    return i // 10
df['AgeState'] = df['Customer_Age'].apply(addColumnAgeState)
df.groupby('AgeState').size()
```

```python
# Q60. Education_Level의 값중 Graduate단어가 포함되는 값은 1 그렇지 않은 경우에는
# 0으로 변경하여 newEduLevel 컬럼을 정의하고 빈도수를 출력하라

def addNewEduLevel(s):
    if 'Graduate' in s:
        return 1
    return 0
df['newEduLevel'] = df.Education_Level.map(lambda x : 1 if 'Graduate' in x else 0)
df['newEduLevel'] = df['Education_Level'].apply(addNewEduLevel)
df.groupby('newEduLevel').size()
```

```python
# Q61. Credit_Limit 컬럼값이 4500 이상인 경우 1 그외의 경우에는 모두 0으로 하는 newLimit 정의하라.
# newLimit 각 값들의 빈도수를 출력하라

df['newLimit'] = df['Credit_Limit'].apply(lambda x: 1 if x >= 4500 else 0)
df.groupby('newLimit').size()
```

```python
# Q62. Marital_Status 컬럼값이 Married 이고 Card_Category 컬럼의 값이 Platinum인 경우 1 그외의 경우에는 ★
# 모두 0으로 하는 newState컬럼을 정의하라. newState의 각 값들의 빈도수를 출력하라 

def addNewState(x):
    if x[0] == 'Married' and x[1] == 'Platinum':
        return 1
    return 0
df['newState'] = df[['Marital_Status', 'Card_Category']].apply(addNewState, axis=1)
df['newState'].value_counts()

def check(x):
    if x.Marital_Status =='Married' and x.Card_Category =='Platinum':
        return 1
    return 0
df['newState'] = df.apply(check,axis=1)
df['newState'].value_counts()
```

```python
# Q67. Yr_Mo_Dy에 년도가 2061년 이상의 경우에는 모두 잘못된 데이터이다.
# 해당경우의 값은 100을 빼서 새롭게 날짜를 Yr_Mo_Dy 컬럼에 정의하라


def newYr_Mo_Dy(x):
    import datetime
    year = x.year - 100 if x.year >= 2061 else x.year
    return pd.to_datetime(datetime.date(year, x.month, x.day))
df['Yr_Mo_Dy'] = df['Yr_Mo_Dy'].apply(newYr_Mo_Dy)
df['Yr_Mo_Dy']
```
